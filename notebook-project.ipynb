**Menangani Outlier**

````python
# menangani outlier pada customer_zip_code_prefix
Q1 = customers_df['customer_zip_code_prefix'].quantile(0.25)
Q3 = customers_df['customer_zip_code_prefix'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = customers_df[(customers_df['customer_zip_code_prefix'] < lower_bound) | (customers_df['customer_zip_code_prefix'] > upper_bound)]
print(outliers)
````

**Insight**
- Tidak ada outlier pada table customers_df kolom customer_zip_code_prefix

````python
# menangani outlier pada geolocation_lat
Q1 = geolocation_df['geolocation_lat'].quantile(0.25)
Q3 = geolocation_df['geolocation_lat'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = geolocation_df[(geolocation_df['geolocation_lat'] < lower_bound) | (geolocation_df['geolocation_lat'] > upper_bound)]
print(outliers)
````

**Insight**
- Outlier latitude yang teridentifikasi mencakup beberapa lokasi di berbagai kota dan negara bagian di Brazil, seperti SP (São Paulo) dan RS (Rio Grande do Sul).

````python
# Menghilangkan outlier dengan filter data (menyimpan hanya nilai-nilai yang berada dalam range IQR)
# Menghitung Q1, Q3, dan IQR
Q1 = geolocation_df['geolocation_lat'].quantile(0.25)
Q3 = geolocation_df['geolocation_lat'].quantile(0.75)
IQR = Q3 - Q1

# Menghitung batas bawah dan atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Menghapus outlier
geolocation_df_clean = geolocation_df[(geolocation_df['geolocation_lat'] >= lower_bound) & (geolocation_df['geolocation_lat'] <= upper_bound)]

# Cek apakah masih ada outlier
outliers = geolocation_df_clean[(geolocation_df_clean['geolocation_lat'] < lower_bound) | (geolocation_df_clean['geolocation_lat'] > upper_bound)]

# Menampilkan hasil
print(outliers)
````

**Insight**
- Sudah tidak ada outlier pada kolom geolocation_lat
